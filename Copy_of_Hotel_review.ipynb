```python
import pandas as pd
from transformers import pipeline
import datetime
```

```python
# Load sentiment pipeline
sentiment_model = pipeline("sentiment-analysis")
```

```python
# Collect inputs
review_text = input("Enter your review: ")

# Structured ratings
rating = int(input("Overall rating (1-5): "))
room_clean = int(input("Room cleanliness (1-5): "))
food_quality = int(input("Food quality (1-5): "))
staff_behavior = int(input("Staff behavior (1-5): "))
room_service = int(input("Room service (1-5): "))
location = int(input("Location convenience (1-5): "))
amenities = int(input("Amenities (Wi-Fi, pool, etc.) (1-5): "))
value_money = int(input("Value for money (1-5): "))

# Stay metadata
duration = input("Duration of stay (e.g., 3 nights): ")
room_type = input("Type of room (Deluxe/Suite/etc.): ")
visit_purpose = input("Purpose of visit (Business/Leisure): ")
booking_source = input("Booking source (App/Website/Agency): ")

# Time
submitted_at = datetime.datetime.now()
```

```python
# Run sentiment analysis on the review
sentiment = sentiment_model(review_text)[0]

# Show output
print(f"\nPredicted Sentiment: {sentiment['label']} (Score: {sentiment['score']:.2f})")
```

```python
# Combine all data
data = {
    "Review": [review_text],
    "Predicted Sentiment": [sentiment['label']],
    "Sentiment Score": [sentiment['score']],
    "Rating": [rating],
    "Room Cleanliness": [room_clean],
    "Food Quality": [food_quality],
    "Staff Behavior": [staff_behavior],
    "Room Service": [room_service],
    "Location": [location],
    "Amenities": [amenities],
    "Value for Money": [value_money],
    "Stay Duration": [duration],
    "Room Type": [room_type],
    "Visit Purpose": [visit_purpose],
    "Booking Source": [booking_source],
    "Submission Time": [submitted_at]
}

df = pd.DataFrame(data)

# Save to CSV
df.to_csv("hotel_review_output.csv", index=False)
print("\nReview saved to 'hotel_review_output.csv'")
df
```

```python
# Generate rating summary in natural language
def generate_rating_summary(data):
    summary = []

    # Helper function to categorize scores
    def get_feedback(category, score):
        if score >= 4:
            return f"{category} was excellent"
        elif score == 3:
            return f"{category} was average"
        else:
            return f"{category} needs improvement"

    # Overall impression
    if data['Rating'][0] >= 4:
        summary.append("Overall, you had a great experience at this hotel.")
    elif data['Rating'][0] == 3:
        summary.append("Overall, you had a decent experience at this hotel.")
    else:
        summary.append("Overall, you were disappointed with this hotel.")

    # Positive aspects (scores >=4)
    positives = []
    if data['Room Cleanliness'][0] >= 4: positives.append("room cleanliness")
    if data['Food Quality'][0] >= 4: positives.append("food quality")
    if data['Staff Behavior'][0] >= 4: positives.append("friendly staff")
    if data['Room Service'][0] >= 4: positives.append("efficient room service")
    if data['Location'][0] >= 4: positives.append("great location")
    if data['Amenities'][0] >= 4: positives.append("good amenities")
    if data['Value for Money'][0] >= 4: positives.append("good value for money")

    if positives:
        summary.append(f"You particularly appreciated the {', '.join(positives)}.")

    # Areas for improvement (scores <=2)
    improvements = []
    if data['Room Cleanliness'][0] <= 2: improvements.append("room cleanliness")
    if data['Food Quality'][0] <= 2: improvements.append("food quality")
    if data['Staff Behavior'][0] <= 2: improvements.append("staff behavior")
    if data['Room Service'][0] <= 2: improvements.append("room service")
    if data['Location'][0] <= 2: improvements.append("location convenience")
    if data['Amenities'][0] <= 2: improvements.append("amenities")
    if data['Value for Money'][0] <= 2: improvements.append("value for money")

    if improvements:
        summary.append(f"However, the {', '.join(improvements)} could be improved.")

    # Neutral aspects (score =3)
    neutrals = []
    if data['Room Cleanliness'][0] == 3: neutrals.append("room cleanliness was adequate")
    if data['Food Quality'][0] == 3: neutrals.append("food was average")
    if data['Staff Behavior'][0] == 3: neutrals.append("staff were okay")
    if data['Room Service'][0] == 3: neutrals.append("room service was acceptable")
    if data['Location'][0] == 3: neutrals.append("location was neither good nor bad")
    if data['Amenities'][0] == 3: neutrals.append("amenities were standard")
    if data['Value for Money'][0] == 3: neutrals.append("value was fair")

    if neutrals:
        summary.append(f"Some aspects like {', '.join(neutrals)}.")

    return ' '.join(summary)

# Generate and display summary
rating_summary = generate_rating_summary(data)
print("\n--- Rating Summary ---")
print(rating_summary)
```
